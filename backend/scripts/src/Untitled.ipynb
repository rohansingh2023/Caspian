```python
stop_words = set(stopwords.words('english'))
```

```python
def clean_text(text):
    # Tokenize text
    text = str(text)
    tokens = word_tokenize(text.lower())  # Convert to lowercase for uniformity
    # Remove punctuation and stopwords
    tokens = [word for word in tokens if word not in stop_words and word not in string.punctuation]
    return tokens
```

```python
vocabulary = set()
```

```python
def process_array_column(arr):
    words = []
    for item in arr:
        if isinstance(item, str):  # If the item is a string, process it
            words.extend(clean_text(item))
        elif isinstance(item, list):  # If it's already a list of words, clean each element
            for sub_item in item:
                words.extend(clean_text(sub_item))
    return words
```

```python
text_columns = ['title', 'company_name', 'location', 'via',
       'description', 'extensions', 'job_id', 'thumbnail', 'posted_at',
       'schedule_type', 'search_location', 'description_tokens']
```

```python
for col in text_columns:
    for val in df[col].dropna():
        # If the value is a string representation of an array, convert it to a list
        if isinstance(val, str):
            try:
                val = ast.literal_eval(val)  # Convert string representation of list to actual list
            except (ValueError, SyntaxError):
                pass  # If it fails, just treat it as a normal string
        # Now process the value, whether it's a string or a list
        if isinstance(val, list):
            # If it's an array, process each item in the array
            vocabulary.update(process_array_column(val))
        else:
            # If it's just a string, clean the text
            vocabulary.update(clean_text(val))
```

```python
inverted_index = {}
```

```python
# Process each row to build the inverted index
for idx, row in df.iterrows():
    for col in text_columns:
        val = row[col]
        
        if isinstance(val, str):
            # Process string data
            words = clean_text(val)
        elif isinstance(val, list):
            # Process list data (if it's a string representation of a list, convert it first)
            if isinstance(val, str):
                try:
                    val = ast.literal_eval(val)  # Convert string representation of list to actual list
                except (ValueError, SyntaxError):
                    pass  # If it fails, just treat it as a normal string
            words = process_array_column(val)
        
        # Update the inverted index
        for word in words:
            if word not in inverted_index:
                inverted_index[word] = set()  # Initialize as a set to avoid duplicate rows
            inverted_index[word].add(idx)  # Add the row index
```

```python
for word in inverted_index:
    inverted_index[word] = sorted(list(inverted_index[word]))
```
